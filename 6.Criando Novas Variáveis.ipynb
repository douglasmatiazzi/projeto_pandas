## Criando Novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep=';')
```

```python
dados.info()
```

```python
dados.head(10)
```

### Criar Variável Valor Bruto

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head()
```

### Criar Variável Valor/M2

```python
dados['Valor/m²'] = dados['Valor'] / dados['Area']
```

```python
dados.head()
```

```python
dados['Valor/m²'] = dados['Valor/m²'].round(2)
```

```python
dados.head()
```

### Criar Variável Valor Bruto/M2

```python
dados['Valor Bruto/m²'] = dados['Valor Bruto'] / dados['Area']
```

```python
dados['Valor Bruto/m²'] = dados['Valor Bruto/m²'].round(2)
```

```python
dados.head()
```

### Criar Variável Tipo Agregado que junta todos os tipos de Casa

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados.head(10)
```

### Excluindo Variáveis do DataFrame

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor/m²', 'Valor Bruto', 'Valor Bruto/m²']])
```

```python
dados_aux.head()
```

### Deletando um campo com com del

```python
del dados_aux['Valor Bruto']
```

```python
dados_aux.head()
```

### Deletando 2 campos com drop

```python
dados.drop(['Valor Bruto', 'Valor Bruto/m²'], axis=1, inplace=True)
```

```python
dados.head()
```

```python
dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```
